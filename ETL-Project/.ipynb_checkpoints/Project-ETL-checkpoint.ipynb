```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "oscars.csv"
oscars_df = pd.read_csv(csv_file)
oscars_df
```

```python
oscars_df_drop=oscars_df.drop(columns=['Ceremony','Name'])
```

```python
oscars_new_df=oscars_df_drop.loc[oscars_df_drop['Year'] >= 1995]
```

```python
oscars_new_df
```

```python
oscars=oscars_new_df.dropna()
```

```python
oscars_winners=oscars.drop(columns=['Winner'])
```

```python
oscars_winners
```

```python
database = ""
server = "localhost:5432"
engine = create_engine('postgres://postgres:password@localhost:5432/database')
conn = engine.connect()
```

```python
oscars_winners.to_sql(name='oscars', conn=engine, if_exists='append', index=True)
```
